```python
import plotly
import pandas as pd
import numpy as np
from plotly.graph_objs import *
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode()

dist_duration = pd.read_csv("../data/mikeCTsAvgLatLonDist.csv", nrows=10000)
print dist_duration.shape
dist_duration.head(1)
```

```python
binned_travel_times = dist_duration.copy()
binned_travel_times['velocity'] = binned_travel_times['distance']/binned_travel_times['duration']
# binned_travel_times.describe()
# binned_travel_times_grouped = binned_travel_times.groupby(['start_censustract', "end_censustract"], as_index=False).mean()
# binned_travel_times_grouped.index = binned_travel_times_grouped['start_censustract']
# # # pd.cut(binned_travel_times_grouped["distance"], np.arange(0, 23, 2))
# binned_travel_times_grouped_final = binned_travel_times_grouped.groupby(pd.cut(binned_travel_times_grouped["distance"], np.arange(0, 23, 2)), as_index=False).mean()
# binned_travel_times_grouped_final.head(20)

# df = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
#                    'B' : np.random.randn(12)})
# bins = {'one': (-10,-1,0,1,10), 'two':(-100,0,100), 'three':(-999,0,1,2,3)}


binned_travel_times.dropna(axis=0, subset=['start_censustract'], inplace=True)
start_census_tracts = binned_travel_times.start_censustract.astype(int)
start_census_tracts = start_census_tracts.astype(str)
dict_cTs = dict(zip(start_census_tracts.values, start_census_tracts.values))
bins = dict_cTs.fromkeys(dict_cTs, tuple(np.arange(0, 23, 2)))
bins

# # def func(row):
# #     return pd.cut([row['B']], bins=bins[row['A']])[0]

# # df['C'] = df.apply(func, axis=1)

def binize2(df):
#     df['distance bins'] = ''
    for key, values in bins.items():
        mask = df['start_censustract'] == key
        df.loc[mask, 'distance_bins'] = pd.cut(df.loc[mask, 'distance'], bins=values)
        
binned_travel_times.start_censustract = binned_travel_times.start_censustract.astype(int)
binned_travel_times.start_censustract = binned_travel_times.start_censustract.astype(str)
binize2(binned_travel_times)
```

```python
binned_final = binned_travel_times[['duration', 'start_censustract', 'distance', 'velocity', 'distance bins']]
binned_final['CT_Bin'] = binned_final['start_censustract']+'_'+binned_final['distance bins']
print binned_final.shape
binned_final.head(1)
```

```python
binned_final = binned_final.groupby(['start_censustract', "distance bins"], as_index=False).mean()
```

```python
dist_duration['velocity'] = dist_duration['distance']/dist_duration['duration']
de_duped_dist_duration = dist_duration[dist_duration['start_censustract'] != dist_duration['end_censustract']]
grouped = de_duped_dist_duration.groupby(['start_censustract', "end_censustract"], as_index=False).mean()
grouped[['start_censustract', 'end_censustract']] = grouped[['start_censustract', 'end_censustract']].astype(int)
grouped[['start_censustract', 'end_censustract']] = grouped[['start_censustract', 'end_censustract']].astype(str)
grouped_dropped = grouped.drop(['end_censustract', 'start_lat', 'start_long', 'end_lat', 'end_long'], axis=1)
print grouped_dropped.shape
grouped_dropped.head(1)
```

```python
score = grouped_dropped.groupby('start_censustract', as_index=False).mean()
score.head(1)
```

```python
censusdata = pd.read_csv('../data/finalround/census_all.csv', dtype=str)
censusdata = censusdata[['CT2010', 'BoroCode', 'mean_income', 'avg_travel_time', 'median_income', 'poverty_perc']]
censusdata['censustract'] = censusdata['BoroCode']+censusdata['CT2010']
censusdata = censusdata[['BoroCode','mean_income', 'censustract', 'avg_travel_time', 'median_income', 'poverty_perc']]
censusdata['BoroCode'] = censusdata.BoroCode.map({'1':"Manhattan", '2':"Bronx", '3':"Brooklyn", '4':"Queens", '5':"Staten Island"})
censusdata.head(2)
```

```python
cen_merged = score.merge(censusdata, how='left', left_on='start_censustract', right_on='censustract')
cen_merged['mean_income'].astype(float)
cen_merged = cen_merged[['censustract', 'duration', 'distance', 'velocity', 'BoroCode', 'mean_income', 'avg_travel_time', 'median_income', 'poverty_perc',]]
cen_merged.to_csv('ctsumstats.csv')
newf = cen_merged.pivot(columns='BoroCode')
newf.columns = ['_'.join(col).strip() for col in newf.columns.values]
newf.to_csv('ctsumstatspivottable.csv')
newf.head(1)
```

# Correlation Analysis

```python
import itertools
compare_list = ['median_income', 'mean_income', 'poverty_perc', 'distance', 'velocity', 'avg_travel_time', 'duration']
pairs = list(itertools.combinations(compare_list, 2))
for i in pairs:
    print "%s correlated to %s\n" % i, cen_merged[i[0]].corr(cen_merged[i[1]], method='spearman')
```
