```python
import plotly
import pandas as pd
import numpy as np
from plotly.graph_objs import *
from plotly.offline import init_notebook_mode, iplot

dist_duration = pd.read_csv("../data/mikeCTsAvgLatLonDist.csv", nrows=100)
dist_duration
```

```python
dist_duration['transit_index'] = dist_duration['duration'].pow(4)/dist_duration['distance']
de_duped_dist_duration = dist_duration[dist_duration['start_censustract'] != dist_duration['end_censustract']]
grouped = de_duped_dist_duration.groupby(['start_censustract', "end_censustract"], as_index=False).mean()
grouped[['start_censustract', 'end_censustract']] = grouped[['start_censustract', 'end_censustract']].astype(int)
grouped[['start_censustract', 'end_censustract']] = grouped[['start_censustract', 'end_censustract']].astype(str)
print grouped.shape
grouped.head(1)
```

```python
# group by the Start CT value
start_CT_score = grouped.groupby('start_censustract', as_index=False).mean()
print start_CT_score.shape
start_CT_score.rename(columns = {'start_censustract':'censustract'}, inplace=True)

# group by the End CT value
end_CT_score = grouped.groupby('end_censustract', as_index=False).mean()
end_CT_score.rename(columns = {'end_censustract':'censustract'}, inplace=True)
frames = [start_CT_score, end_CT_score]
merged = pd.concat(frames)
merged['text'] = 'duration: '+merged['duration'].astype(str)+'<br>distance: '+merged['distance'].astype(str)+'<br>transit index: '+merged['transit_index'].astype(str)
print merged.shape
merged.head(1)
```

```python
censusdata = pd.read_csv('../data/census_data/documents-export-2016-05-10/census_merged.csv', dtype=str)
censusdata = censusdata[['CT2010', 'BoroCode', 'mean_income', 'avg_travel_time']]
censusdata['censustract'] = censusdata['BoroCode']+censusdata['CT2010']
censusdata = censusdata[['BoroCode','mean_income', 'censustract', 'avg_travel_time']]
censusdata['BoroCode'] = censusdata.BoroCode.map({'1':"Manhattan", '2':"Bronx", '3':"Brooklyn", '4':"Queens", '5':"Staten Island"})
censusdata.head(2)
```

```python
cen_merged = merged.merge(censusdata, how='left', on='censustract')
cen_merged['mean_income'].astype(float)
cen_merged = cen_merged[['censustract', 'duration', 'distance', 'transit_index', 'text', 'BoroCode', 'mean_income', 'avg_travel_time']]
cen_merged.to_csv('ctsumstats.csv')
newf = cen_merged.pivot(columns='BoroCode')
newf.columns = ['_'.join(col).strip() for col in newf.columns.values]
newf.to_csv('ctsumstatspivottable.csv')
```
